## ðŸŒŠ Hydrological Dataset Exploration

In this section, we load and visualize time series data from 4 river monitoring stations:
- River level (`wlvalue`) in meters
- Flow rate (`flvalue`) in mÂ³/s

Our objectives here are to:
- Visually inspect long-term and short-term dynamics
- Detect missing values, seasonal trends, and abrupt events like floods
- Understand the temporal structure of the data before modeling

This exploratory step helps us determine which features are relevant and how to prepare the data for prediction tasks.

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

# List of all four CSVs (assumed to live in your working directory or specify full paths)
file_paths = [
    'Otewa.csv',
    'Hamilton.csv',
    'Te Aroha.csv',
    'Paeroa.csv'
]

# Common plotting colors
color1 = 'tab:blue'
color2 = 'tab:green'

# Processing each dataset in turn
for file_path in file_paths:
    # Derive a simple name for printing/plot titles
    dataset_name = file_path.replace('.csv', '')
    
    # 1. Read CSV (skip title row), keep only needed columns
    df = pd.read_csv(
        file_path,
        skiprows=1,
        usecols=['date', 'time', 'wlvalue', 'flvalue']
    )
    
    # 2. Combine date + time into datetime, set as index
    df['datetime'] = pd.to_datetime(df['date'] + ' ' + df['time'], dayfirst=True)
    df.drop(['date', 'time'], axis=1, inplace=True)
    df.set_index('datetime', inplace=True)
    
    # 3. Rename columns
    df.rename(columns={'wlvalue': 'water_level', 'flvalue': 'flow'}, inplace=True)
    
    # 4. Overview & descriptive stats
    print(f"\n=== Dataset: {dataset_name} ===")
    print("Dimensions:", df.shape)
    print("Data types:")
    print(df.dtypes, "\n")
    print("Missing values per column:")
    print(df.isna().sum(), "\n")
    print("Descriptive statistics:")
    display(df.describe())
    display(df.head())
    
    # 5. Plot water level & flow on two y-axes
    fig, ax1 = plt.subplots(figsize=(12, 5))
    ax1.set_title(f"Time Series: Water Level & Flow ({dataset_name})")
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Water Level', color=color1)
    ax1.plot(df.index, df['water_level'], color=color1, label='Water Level', linewidth=1)
    ax1.tick_params(axis='y', labelcolor=color1)
    
    ax2 = ax1.twinx()
    ax2.set_ylabel('Flow', color=color2)
    ax2.plot(df.index, df['flow'], color=color2, label='Flow', linewidth=1, alpha=0.6)
    ax2.tick_params(axis='y', labelcolor=color2)
    
    fig.tight_layout()
    plt.show()
    
    # 6. Compute 7-day rolling means
    window = 30
    df['wl_roll7'] = df['water_level'].rolling(window).mean()
    df['fl_roll7'] = df['flow'].rolling(window).mean()
    
    # 7. Plot raw vs rolling means on two y-axes
    fig, ax1 = plt.subplots(figsize=(12, 5))
    ax1.set_title(f"Raw vs 7-Day MA ({dataset_name})")
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Raw Water Level', color=color1)
    ax1.plot(df.index, df['water_level'], color=color1, alpha=0.3, label='Raw WL')
    ax1.plot(df.index, df['wl_roll7'],   color=color1, linewidth=2, label='WL 7-day MA')
    ax1.tick_params(axis='y', labelcolor=color1)
    
    ax2 = ax1.twinx()
    ax2.set_ylabel('Raw Flow / 7-Day MA', color=color2)
    ax2.plot(df.index, df['flow'],   color=color2, alpha=0.3, label='Raw Flow')
    ax2.plot(df.index, df['fl_roll7'], color=color2, linewidth=2, label='Flow 7-day MA')
    ax2.tick_params(axis='y', labelcolor=color2)
    
    fig.tight_layout()
    plt.show()
```

## ðŸ§  Classical LSTM Baseline Model

We train a standard **Long Short-Term Memory (LSTM)** network to predict the **next day's river level** (`wlvalue_{t+1}`), using a **365-day sliding window** as input. Each input vector includes:
- Daily river level (`water_level`)
- Daily flow rate (`flow`)
- Cyclical encoding of the day-of-year (`sin_doy`, `cos_doy`)
- 7-day moving average of river level (`mov_avg`)

### ðŸ”§ Model Overview:
- Features are normalized using `MinMaxScaler`
- The LSTM is trained on the full history except for the final 60 days
- Evaluation is performed on the last 60 days to assess generalization
- A Dense layer maps the final hidden state to the predicted `wlvalue_{t+1}`

This model acts as our **classical baseline** for comparison against the hybrid quantum-enhanced version.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense

# -----------------------------------
# 1. LOAD & CLEAN DATA
# -----------------------------------
file_path = 'Otewa.csv'

# Read CSV: skip title + blank line, read only needed columns
df = pd.read_csv(
    file_path,
    skiprows=2,
    usecols=['date', 'time', 'wlvalue', 'flvalue']
)

# Parse datetime, coerce bad formats to NaT
df['datetime'] = pd.to_datetime(
    df['date'] + ' ' + df['time'],
    dayfirst=True,
    errors='coerce'
)

# Convert wlvalue and flvalue to floats, coercing bad data to NaN
df['wlvalue'] = pd.to_numeric(df['wlvalue'], errors='coerce')
df['flvalue'] = pd.to_numeric(df['flvalue'], errors='coerce')

# Drop any rows where we failed to parse datetime or numeric values
df.dropna(subset=['datetime', 'wlvalue', 'flvalue'], inplace=True)

# Set datetime index and sort to guarantee monotonicity
df.drop(['date', 'time'], axis=1, inplace=True)
df.set_index('datetime', inplace=True)
df.sort_index(inplace=True)

# Rename for clarity
df.rename(columns={'wlvalue': 'water_level', 'flvalue': 'flow'}, inplace=True)

# -----------------------------------
# 2. FEATURE ENGINEERING & SPLIT
# -----------------------------------
# Add cyclical day-of-year features to capture annual seasonality
doy = df.index.dayofyear.values
df['doy_sin'] = np.sin(2 * np.pi * doy / 365.25)
df['doy_cos'] = np.cos(2 * np.pi * doy / 365.25)

# Stack into a NumPy array
features = ['water_level', 'flow', 'doy_sin', 'doy_cos']
data = df[features].values
n_features = data.shape[1]

# Define test period (last 60 days) and window size (365 days)
test_days = 60
window_size = 365

# Split so that test_data starts window_size days before the end
split_idx = len(data) - test_days
train_data = data[:split_idx]
test_data  = data[split_idx - window_size:]

# Scale based on training data only
scaler = MinMaxScaler()
scaler.fit(train_data)
train_scaled = scaler.transform(train_data)
test_scaled  = scaler.transform(test_data)

# Function to build sliding-window sequences
def create_sequences(arr, window):
    X, y = [], []
    for i in range(len(arr) - window):
        X.append(arr[i : i + window])
        y.append(arr[i + window, 0])  # next-day water_level
    return np.array(X), np.array(y)

# Create train/test sequences
X_train, y_train = create_sequences(train_scaled, window_size)
X_test,  y_test  = create_sequences(test_scaled,  window_size)

# -----------------------------------
# 3. BUILD & TRAIN LSTM MODEL
# -----------------------------------
model = Sequential([
    LSTM(64, input_shape=(window_size, n_features)),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse')

history = model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=32,
    validation_data=(X_test, y_test)
)

# -----------------------------------
# 4. PREDICT & INVERSE TRANSFORM
# -----------------------------------
y_pred_scaled = model.predict(X_test)

# Prepare zeros to pad back to full feature dimension
pad = np.zeros((len(y_pred_scaled), n_features - 1))

# Inverse-transform predictions and true values
pred_full = scaler.inverse_transform(np.hstack([y_pred_scaled, pad]))
true_full = scaler.inverse_transform(np.hstack([y_test.reshape(-1, 1), pad]))

y_pred = pred_full[:, 0]
y_true = true_full[:, 0]

# -----------------------------------
# 5. PLOT RESULTS
# -----------------------------------
dates_test = df.index[split_idx : split_idx + len(y_true)]

plt.figure(figsize=(12, 6))
plt.plot(dates_test, y_true, label='Actual Water Level')
plt.plot(dates_test, y_pred, label='Predicted Water Level', alpha=0.8)
plt.title('Next-Day Water Level: Actual vs. Predicted (Last 60 Days)')
plt.xlabel('Date')
plt.ylabel('Water Level')
plt.legend()
plt.tight_layout()
plt.show()
```

## âš›ï¸ QLSTM with QRU (Quantum Re-uploading Unit)

This section introduces a **quantum-classical hybrid model**, combining:
- LSTM-style temporal modeling
- A **QRU (Quantum Re-uploading Unit)** inserted into the **candidate gate** of the memory cell

### ðŸš€ Why use a QRU?
The QRU is a small quantum circuit with **5 qubits**, each encoding one of the following features:
- `water_level_t`, `flow_t`
- `sin_doy`, `cos_doy` (to model seasonality)
- `mov_avg` (to model local trend)

The circuit uses **3 layers of data re-uploading**, which allows it to:
- Efficiently capture **both low- and high-frequency patterns**
- Learn complex nonlinear structures in the data
- Improve precision for sudden events (e.g., floods) and periodic signals

### ðŸ§  Architecture Details:
- Forget, input, and output gates are classical (Dense + sigmoid)
- Candidate gate is replaced by a **QRU circuit output**
- The quantum output is projected into the LSTM hidden dimension

Training and evaluation follow the same structure as the classical LSTM to ensure a fair and interpretable performance comparison.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import pennylane as qml
import tensorflow as tf
from tensorflow.keras.layers import Layer, Dense, Input, RNN
from tensorflow.keras.models import Model

# -----------------------------------
# 1. LOAD & CLEAN DATA
# -----------------------------------
file_path = 'Otewa.csv'

df = pd.read_csv(
    file_path,
    skiprows=2,
    usecols=['date','time','wlvalue','flvalue']
)
df['datetime'] = pd.to_datetime(df['date'] + ' ' + df['time'], dayfirst=True, errors='coerce')
df['wlvalue'] = pd.to_numeric(df['wlvalue'], errors='coerce')
df['flvalue'] = pd.to_numeric(df['flvalue'], errors='coerce')
df.dropna(subset=['datetime','wlvalue','flvalue'], inplace=True)
df.drop(['date','time'], axis=1, inplace=True)
df.set_index('datetime', inplace=True)
df.sort_index(inplace=True)
df.rename(columns={'wlvalue':'water_level','flvalue':'flow'}, inplace=True)

# -----------------------------------
# 2. FEATURE ENGINEERING
# -----------------------------------
doy = df.index.dayofyear.values
df['doy_sin'] = np.sin(2*np.pi*doy/365.25)
df['doy_cos'] = np.cos(2*np.pi*doy/365.25)
df['mov_avg'] = df['water_level'].rolling(7).mean()
df.dropna(inplace=True)

# -----------------------------------
# 3. PREPARE DATA
# -----------------------------------
features = ['water_level','flow','doy_sin','doy_cos','mov_avg']
data = df[features].values
n_features = data.shape[1]

test_days = 60
window_size = 365
split_idx = len(data) - test_days
train_data = data[:split_idx]
test_data  = data[split_idx - window_size:]

scaler = MinMaxScaler()
scaler.fit(train_data)
train_scaled = scaler.transform(train_data)
test_scaled  = scaler.transform(test_data)

def create_sequences(arr, window):
    X, y = [], []
    for i in range(len(arr) - window):
        X.append(arr[i:i+window])
        y.append(arr[i+window, 0])
    return np.array(X), np.array(y)

X_train, y_train = create_sequences(train_scaled, window_size)
X_test, y_test = create_sequences(test_scaled, window_size)

# -----------------------------------
# 4. QUANTUM CIRCUIT
# -----------------------------------
n_qubits = 5
depth = 3
dev = qml.device('default.qubit', wires=n_qubits)

@qml.qnode(dev, interface='tf')
def qru(inputs, weights):
    for d in range(depth):
        for i in range(n_qubits):
            qml.RY(inputs[i], wires=i)
        for i in range(n_qubits):
            qml.RZ(weights[d, i], wires=i)
        for i in range(n_qubits-1):
            qml.CRX(weights[d, i], wires=[i, i+1])
        qml.CRX(weights[d, n_qubits-1], wires=[n_qubits-1, 0])
    return qml.expval(qml.PauliZ(0))

# -----------------------------------
# 5. QUANTUM CANDIDATE GATE
# -----------------------------------
class QuantumCandidateGate(Layer):
    def __init__(self, hidden_dim):
        super().__init__()
        initializer = tf.random_uniform_initializer(0, 2*np.pi)
        self.q_weights = tf.Variable(initializer((depth, n_qubits)), trainable=True)
        self.proj = Dense(hidden_dim)

    def call(self, inputs):
        q_out = tf.vectorized_map(lambda x: qru(x, self.q_weights), inputs)
        q_out = tf.expand_dims(q_out, -1)
        return self.proj(q_out)

# -----------------------------------
# 6. QLSTM CELL
# -----------------------------------
class QLSTMCell(Layer):
    def __init__(self, hidden_dim):
        super().__init__()
        self.hidden_dim = hidden_dim
        self.state_size = [hidden_dim, hidden_dim]
        self.output_size = hidden_dim
        self.c_gate = QuantumCandidateGate(hidden_dim)
        self.f_gate = Dense(hidden_dim, activation='sigmoid')
        self.i_gate = Dense(hidden_dim, activation='sigmoid')
        self.o_gate = Dense(hidden_dim, activation='sigmoid')

    def call(self, inputs, states):
        h_prev, c_prev = states
        gate_in = tf.concat([inputs, h_prev], axis=-1)
        f_t = self.f_gate(gate_in)
        i_t = self.i_gate(gate_in)
        o_t = self.o_gate(gate_in)
        c_tilda = self.c_gate(inputs)
        c_t = f_t * c_prev + i_t * c_tilda
        h_t = o_t * tf.tanh(c_t)
        return h_t, [h_t, c_t]

# -----------------------------------
# 7. BUILD & TRAIN MODEL
# -----------------------------------
hidden_dim = 32
cell = QLSTMCell(hidden_dim)
rnn_layer = RNN(cell, return_sequences=False)
inputs = Input(shape=(window_size, n_features))
x = rnn_layer(inputs)
outputs = Dense(1)(x)
model = Model(inputs, outputs)
model.compile(optimizer='adam', loss='mse')

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=20,
    batch_size=16
)

# -----------------------------------
# 8. PREDICT & PLOT
# -----------------------------------
y_pred_scaled = model.predict(X_test)
pad = np.zeros((len(y_pred_scaled), n_features-1))
pred_full = scaler.inverse_transform(np.hstack([y_pred_scaled, pad]))
true_full = scaler.inverse_transform(np.hstack([y_test.reshape(-1,1), pad]))
y_pred = pred_full[:,0]
y_true = true_full[:,0]
dates_test = df.index[split_idx:split_idx + len(y_true)]

plt.figure(figsize=(12,6))
plt.plot(dates_test, y_true, label='Actual WL')
plt.plot(dates_test, y_pred, label='Predicted WL', alpha=0.8)
plt.title('QLSTM-QRU Next-Day Water Level (Last 60 Days)')
plt.xlabel('Date')
plt.ylabel('Water Level')
plt.legend()
plt.tight_layout()
plt.show()
```
